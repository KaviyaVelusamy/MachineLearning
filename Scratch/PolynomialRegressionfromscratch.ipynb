```python
"""LinearRegression
It works only on linear data
Linear regression-linear polynomial -degree1
Quadratic Polynomial-polynomial with degree 2(ax^2+bx+c)
Cubic polynomail-polynomial with degree 3(ax^3+bX^2+cx+d)
parameter-data given to train the model
Hyperparameter-used to tune the model
Forward selection:increase the degree parameter till you get the opyimal results
Backward selection:
Why is polynomial regression is called linear regression:we could convert polynomial to linear polynomial
What happens if we increase the degree more higher?-overfit(good during training but not during testing)(it memorize values in a wrong answer model)
Loss and cost function-using only one data to get error and taking all datas to calculate the error){summation(i=1to n)(yi-yi')^2/n}{summation(i=1to n)(yi-yi')^2}
Bias-error(difference between predicted value and actual value)-calculates using simple features-these assumptions are always not correct
Variance-(split of data according to mean)(mugging up if large number of features)-takes into account the fluctuations in the data.i.e.,noise as well-model learns too much from the data

High bias low variance -underfitting
Low bias high variance-overfitting"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([23,43,45,67,89])
```

```python
degree=2
```

```python
x_poly=np.column_stack([x**i for i in range(degree+1)])
x_poly
```

```python
weights=np.zeros(degree+1)
weights
```

```python
#epochs~=iterations
```

```python
learning_rate=0.001
num_iterations=10000
```

```python
for _ in range(num_iterations):
  #calculate predictions
  predictions=np.dot(x_poly,weights)
  #calculate the error
  error=predictions-y
  #updates the weights(coefficients)using gradient descent
  gradient=np.dot(x_poly.T,error)/len(x)
  weights-=learning_rate*gradient
  print(f'weights for iteration {_}:{weights}')
```

```python
def predict(x,weights):
  x_poly=np.array([x**i for i in range(degree+1)])
  return np.dot(x_poly,weights)
```

```python
plt.scatter(x,y,label="Data points")
x_range=np.linspace(min(x),max(x),100)#to space evenly
y_pred=[predict(x,weights) for x in x_range]
plt.plot(x_range,y_pred,color='r',label="Polynomial Regression")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
