```python
data['buying'],_=pd.factorize(data['buying'])#factorize returns the categorical value changed into the numerical value and also returns category,it is one of the feature engineering technique
data['maint'],_=pd.factorize(data['maint'])#factorize -label encoding
data['doors'],_=pd.factorize(data['doors'])
data['persons'],_=pd.factorize(data['persons'])
data['lug_boot'],_=pd.factorize(data['lug_boot'])
data['safety'],_=pd.factorize(data['safety'])
data
```

```python
x=data.iloc[:,0:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)#default random_state=42 that is called life number,random_state=0 no mixing of data for training and testing ffirst 70% for training and the remaining 30 is for testing
```

```python
x.columns
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
y_train_pred=dtree.predict(x_train)
```

```python
accuracy=metrics.accuracy_score(y_train,y_train_pred)
print('Train accuracy: {:.2f}'.format(accuracy))
```

```python
y_test_pred=dtree.predict(x_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_test_pred)
print('Train accuracy: {:.2f}'.format(accuracy))
```
